```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import sklearn  
#cross validation to split data into training and testing sets
from sklearn.cross_validation import train_test_split
```

```python
df=pd.read_csv('Classified data.csv')
```

```python
df.head()
```

```python
#standardizing to the same scale
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
#USING SCALER OBJECT TO DO TRANSFORMATION
scaled_features=scaler.transform(df.drop('TARGET CLASS',axis=1))
```

```python
scaled_features
```

```python
#using scaled features to create a dataframe
df_feat=pd.DataFrame(scaled_features,columns=df.columns[:-1])
```

```python
df_feat.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df_feat
y=df['TARGET CLASS']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
predictions=knn.predict(X_test)
```

```python
predictions
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```
